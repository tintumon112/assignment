```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r'C:\Users\user\Downloads\USArrests.csv')
```

```python
x=data[['Murder','Assault','UrbanPop', 'Rape']]
```

```python
from sklearn.cluster import KMeans
wcss=[]
for i in range(1,11):
    km=KMeans(n_clusters=i,n_init=10)
    km.fit(x)
    css=km.inertia_
    wcss.append(css)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss,'o-')
plt.xlabel('No of clusters')
plt.ylabel('wcss score')
plt.title('No of clusters vs wcss')
plt.grid()
plt.show()
```

```python
km=KMeans(n_clusters=3,n_init=10)
```

```python
km.fit(x)
```

```python
cluster_labels=km.predict(x)
```

```python
colors=('r','g','b')
for i in range(max(cluster_labels)+1):
    cluster_data=x[cluster_labels==i]
    plt.scatter(cluster_data['Assault'],
                cluster_data['Murder'],
                c=colors[i],
                label=f'Cluster{i}')
    plt.xlabel('uassault')
    plt.ylabel('murder')
    plt.legend()
```

```python
from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import dendrogram
from scipy.cluster.hierarchy import fcluster
```

```python
mergings=linkage(x,method="ward")
```

```python
plt.figure(figsize=(10,6))
dendrogram(mergings)
plt.axhline(350,color='green',linestyle='--')
plt.show()
```

```python
cluster_labels=fcluster(mergings,350,criterion="distance")
```

```python
cluster_labels
```

```python
plt.scatter(x['Assault'],x['Murder'],c=cluster_labels)
plt.xlabel('Assault')
plt.ylabel('Murder')
plt.title('Assault vs Murder')
plt.show()
```
